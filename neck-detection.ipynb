```python
import cv2
import mediapipe as mp
import time
import math
```

```python
def find_distance(x1, y1, x2, y2):
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)
```

```python
def find_angle(x1, y1, x2, y2):
    theta = math.acos((y2-y1)*(-y1) / (math.sqrt((x2-x1)**2 + (y2-y1)**2) * y1))
    degree = int(180/math.pi) * theta

    return degree
```

```python
# Initilize frame counters.
good_frames = 0
bad_frames = 0

# Font type.
font = cv2.FONT_HERSHEY_SIMPLEX

# Colors.
blue = (255, 127, 0)
red = (50, 50, 255)
green = (127, 255, 0)
dark_blue = (127, 20, 0)
light_green = (127, 233, 100)
yellow = (0, 255, 255)
pink = (255, 0, 255)
```

```python
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5, model_complexity=1)
mp_draw = mp.solutions.drawing_utils

cap = cv2.VideoCapture(0)

# Meta
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_size = (width, height)
```

```python
print('Processing..')

if not cap.isOpened():
    print('Cannot open camera.')
    exit()

while True:
    # Capture frames.
    success, image = cap.read()
    if not success:
        print("Error: Failed to capture frame.")
        break
    
    start = time.time()
    
    # Display the frame
    cv2.imshow('Camera Feed', image)

    # Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    # Get height and width.
    h, w = image.shape[:2]

    # Convert the BGR image to RGB.
    imgRBG = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Improve performance
    image.flags.writeable = True

    # Process the image.
    keypoints = pose.process(imgRBG)

    # Use lm and lmPose as representative of the following methods.
    lm = keypoints.pose_landmarks
    lmPose = mp_pose.PoseLandmark

    if lm:
        print(lm)
        mp_draw.draw_landmarks(image=image, landmark_list=lm, landmark_drawing_spec=mp_pose.POSE_CONNECTIONS)
        for i in range(0, 3):
            print(f'{lmPose(i).name}:\n{lm.landmark[lmPose(i).value]}')
        

    
    
print('Finished.')
cap.release()
cv2.destroyAllWindows()
```
